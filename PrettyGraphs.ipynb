```
import numpy as np
import pandas as pd

# Load the data
data = pd.read_csv('plotter.csv')
```

Check (and sort) the substantial difference in parameter size for the pre-trained models

```
data.sort_values(by="Parameters")
```

**Visualize**

```
axx = data.plot(
    kind='barh',
    x='Pre-trained Model',
    y='Parameters',
    figsize=(12,8),
    xlabel = 'Models',
    ylabel = 'Parameters (millions)',
    title = 'Model Size Difference')
```

```
axx = data.plot(
    kind='line',
    x='Pre-trained Model',
    y='Parameters',
    figsize=(12,8),
    xlabel = 'Models',
    ylabel = 'Parameters (millions)',
    title = 'Model Size Difference')

```
