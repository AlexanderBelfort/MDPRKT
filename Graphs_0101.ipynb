```
import numpy as np
import pandas as pd

# Load the data
data = pd.read_csv('plotter.csv')
```

Check (and sort) the substantial difference in parameter size for the pre-trained models

```
data.sort_values(by="Parameters")
```

**Visualize without GPT-2 and GTP-3**

```
# drop all gpts
# data = data[:-12]

axx = data.sort_values(by="Parameters").plot(
    kind='barh',
    x='Pre-trained Model',
    y='Parameters',
    figsize=(12,8),
    xlabel = 'Models',
    ylabel = 'Parameters (millions)',
    title = 'Model Size Difference')
```

```
axx = data.plot(
    kind='line',
    x='Pre-trained Model',
    y='Parameters',
    figsize=(12,8),
    xlabel = 'Models',
    ylabel = 'Parameters (millions)',
    title = 'Model Size Difference')

```
